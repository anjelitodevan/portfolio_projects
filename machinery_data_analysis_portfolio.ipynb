```python
import pandas as pd
import os
import platform
```

#### Import Data

```python
path = r"C:\Users\Anjelito\Documents\analysis - machinery data cleaning\data_cleaning_challenge.csv"
data_import = pd.read_csv(path)
data_import.head()
```

#### Drop Extra Columns

```python
drop_extra = data_import.drop(["Unnamed: 9", "Unnamed: 10"], axis = "columns")
drop_extra.head()
```

#### Drop Nan Rows

```python
drop_nan = drop_extra[drop_extra["Row Type"].notna()]
drop_nan.head()
```

#### Make iterarion to separate the dataframe

```python
column_values = []
counter = 0

for i in drop_nan["Row Type"]:
    if "first name" in i:
        counter = counter + 1
    column_values.append(counter)
```

```python
iter_cols = drop_nan
iter_cols["Iteration"] = column_values
iter_cols.head()
```

```python
drop_extra_column_name_row = iter_cols[iter_cols["Row Type"] != "Row Type"]
drop_extra_column_name_row.head()
```

#### Separate the name dafaframe

```python
name_dataframe = drop_extra_column_name_row[drop_extra_column_name_row["Row Type"].str.contains("first name")]
name_dataframe.drop(name_dataframe.columns[3:8], axis = 1, inplace= True)
name_dataframe.rename(columns= {"Row Type": "First Name", "Iter Number": "Last Name", 
                               "Power1": "Date"}, inplace= True)
name_dataframe["First Name"] = name_dataframe["First Name"].str[12:]
name_dataframe["Last Name"] = name_dataframe["Last Name"].str[11:]
name_dataframe["Date"] = name_dataframe["Date"].str[6:]
name_dataframe.head()
```

#### Separate the no name dafaframe

```python
no_name_dataframe = drop_extra_column_name_row[~drop_extra_column_name_row["Row Type"].str.contains("first name")]
no_name_dataframe.head()
```

#### Merge the dataframes

```python
final_df = pd.merge(left = name_dataframe, right = no_name_dataframe, how = "inner", on = "Iteration")
final_df.head()
```
