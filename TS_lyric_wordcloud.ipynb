```python
import numpy as np
from PIL import Image
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
import pandas as pd
import glob
import os
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
%matplotlib inline
```

```python
path = r"C:\Users\Anjelito\Documents\WORK\analysis - TS lyric\data"       
all_files = glob.glob(os.path.join(path, "*.csv"))
df = pd.concat((pd.read_csv(f) for f in all_files))
```

```python
df_indexed = df.reset_index().drop("index", axis = "columns")
```

```python
df_indexed.head()
```

```python
text = " ".join(lyric for lyric in df_indexed.lyric).lower()
```

```python
print ("There are {} words in the combination of all review.".format(len(text)))
```

```python
stopwords = set(STOPWORDS)
stopwords.update(["oh oh", "oh"])

wordcloud = WordCloud(width= 3000, height= 1800, stopwords=stopwords, 
                      background_color="white", colormap='Accent').generate(text)
```

```python
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```

```python
wordcloud.to_file("output1.png")
```
